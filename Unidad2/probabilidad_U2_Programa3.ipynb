```python
import math as mt
from itertools import permutations
from itertools import product
import numpy as np
```

```python
asistentes = ["Alejandro", "Bianca", "Carlos", "Daniela", "Erick", "Fabian", "Gabriela", "Hugo", "Ivonne", "Javier"]
n = len(asistentes)
r = 4 # ¿Como agrupar?

Pr = mt.factorial(n) / (mt.factorial(n - r))

print("Existen ", Pr , " posibles combinaciones sin repetición del total de ",
n," asistentes ","en grupos de ", r ," en ", r)

print("Aquí las combinaciones")
permutaciones_asistentes = list(permutations(asistentes, r))

# Imprime las permutaciones
for permutacion in permutaciones_asistentes:
    print(permutacion)
```
