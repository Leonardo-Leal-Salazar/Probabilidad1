estima la siguiente distribución de probabilidad:

```python
import math
import numpy as np
import matplotlib.pyplot as plt

n = [0, 1, 2, 3, 4, 5]
p = [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]

suma = sum(p)
print("La suma de las probabilidades es: ", suma)

# Grafica distribucion
plt.stem(n,p)
plt.title("Distribucion de Probabilidad")
plt.xlabel("Calificacion")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()

# inicializar una variable
valor_esperado = 0

# Recorrer los valores y sus probabilidades
for i in range(len(n)):
  ni = n[i] # Valor de la variable aleatoria
  pi = p[i] # Probabilidad de ese valor
  producto = ni * pi # Multiplicar valor por su probabilidad
  valor_esperado += producto # Sumar al acumulador
  # Mostrar cada paso
  print(f"n = {ni}, p = {pi}, n*p = {producto:.2f}")

# mostrar resultado final
print(f"\nValor esperado (esperanza): {valor_esperado}")


```

```python
# inicializar una variable
valor_esperado = 0

# Recorrer los valores y sus probabilidades
for i in range(len(n)):
  ni = n[i] # Valor de la variable aleatoria
  pi = p[i] # Probabilidad de ese valor
  producto = ni * pi # Multiplicar valor por su probabilidad
  valor_esperado += producto # Sumar al acumulador
  # Mostrar cada paso
  print(f"n = {ni}, p = {pi}, n*p = {producto:.2f}")

# mostrar resultado final
print(f"\nValor esperado (esperanza): {valor_esperado}")
```

# 4.- Construya la función de distribución acumulada 𝑭 𝒙 y represéntela gráficamente.

```python
# Calcular funcion acumulada
F = np.cumsum(p)

# Mostrar resultado
for ni, fi in zip(n, F):
    print(f"F({ni}) = {fi:.2f}")

# Graficar
plt.step(n, F, where='post')
plt.title("Funcion de Distribucion Acumulada (FDA)")
plt.xlabel("Calificacion")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```
