• Un investigador biomédico realiza un conteo del número de células anormales
observadas en una muestra de tejido tomada de distintos pacientes.

• La variable aleatoria 𝑿 representa el número de células anormales encontradas en
una muestra seleccionada al azar de una lámina. A partir de observaciones previas, se
estima la siguiente distribución de probabilidad:

Número de
células
anormales (X)   0 1 2 3 4 5

Probabilidad P(X) 0.10 0.25 0.30 0.20 0.10 0.05

1. Verifique que las probabilidades sumen 1.
2. Elabore la distribución de probabilidad de 𝑿.
3. Calcule:

• El valor esperado 𝑬 𝑿 (promedio de células anormales por muestra).

• La varianza y la desviación estándar.

4. Construya la función de distribución acumulada 𝑭 𝒙 y represéntela gráficamente.

5. Interprete los resultados en contexto biomédico:

• ¿Cuál es el promedio esperado de células anormales por muestra?

• ¿Existe mucha variabilidad entre las muestras?

• ¿Qué indica la función acumulada sobre la probabilidad de encontrar un número bajo o alto de células anormales

```python
import math
import numpy as np
import matplotlib.pyplot as plt

n = [0, 1, 2, 3, 4, 5]
p = [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]

# ============================================================
# PUNTO 1: Verificar que las probabilidades sumen 1
# ============================================================

suma = sum(p)
print("La suma de las probabilidades es: ", suma)

```

```python
# ============================================================
# PUNTO 2: Distribución de probabilidad de X
# ============================================================

# Mostrar tabla de distribución
print("Distribución de Probabilidad de X:")
print(f"{'X':<10} {'P(X)':<10}")
print("-" * 20)
for ni, pi in zip(n, p):
    print(f"{ni:<10} {pi:<10.2f}")

# Grafica distribución
plt.stem(n, p)
plt.title("Distribución de Probabilidad de X\n(Células anormales por muestra)")
plt.xlabel("Número de células anormales (X)")
plt.ylabel("Probabilidad P(X)")
plt.grid(True)
plt.show()
```

```python
# ============================================================
# PUNTO 3: Valor esperado, varianza y desviación estándar
# ============================================================

import math

# --- Valor esperado E(X) ---
valor_esperado = 0
print("Cálculo del Valor Esperado E(X):")
print(f"{'X':<10} {'P(X)':<10} {'X·P(X)':<10}")
print("-" * 30)

for i in range(len(n)):
    ni = n[i]
    pi = p[i]
    producto = ni * pi
    valor_esperado += producto
    print(f"{ni:<10} {pi:<10.2f} {producto:<10.2f}")

print(f"\n  E(X) = {valor_esperado:.4f} células anormales")

# --- Varianza Var(X) ---
e_x2 = 0
print("\nCálculo de la Varianza Var(X):")
print(f"{'X':<10} {'P(X)':<10} {'X²·P(X)':<10}")
print("-" * 30)

for ni, pi in zip(n, p):
    producto2 = ni**2 * pi
    e_x2 += producto2
    print(f"{ni:<10} {pi:<10.2f} {producto2:<10.2f}")

varianza = e_x2 - valor_esperado**2
desviacion_std = math.sqrt(varianza)

print(f"\n  E(X²)              = {e_x2:.4f}")
print(f"  Var(X) = E(X²) - [E(X)]² = {e_x2:.4f} - {valor_esperado:.4f}² = {varianza:.4f}")
print(f"  Desviación estándar σ = √{varianza:.4f} = {desviacion_std:.4f} células anormales")
```

```python
# ============================================================
# PUNTO 4: Función de Distribución Acumulada F(x)
# ============================================================

import numpy as np
import matplotlib.pyplot as plt

# Calcular FDA
F = np.cumsum(p)

# Mostrar tabla
print("Función de Distribución Acumulada F(x):")
print(f"{'X':<10} {'P(X)':<10} {'F(X)':<10}")
print("-" * 30)
for ni, pi, fi in zip(n, p, F):
    print(f"{ni:<10} {pi:<10.2f} {fi:<10.2f}")

# Graficar
plt.step(n, F, where='post', color='steelblue', linewidth=2, label='F(x)')
plt.plot(n, F, 'o', color='steelblue', markersize=7)
plt.title("Función de Distribución Acumulada F(x)\n(Células anormales por muestra)")
plt.xlabel("Número de células anormales (X)")
plt.ylabel("Probabilidad acumulada F(X)")
plt.grid(True)
plt.ylim(0, 1.05)
plt.legend()
plt.show()
```

# 5. Interprete los resultados en contexto biomédico:
## • ¿Cuál es el promedio esperado de células anormales por muestra?
### El valor esperado es E(X) = 2.10 células anormales por muestra. Esto significa que si el investigador tomara muchas muestras de tejido de distintos pacientes, en promedio encontraría alrededor de 2 células anormales por lámina.

## • ¿Existe mucha variabilidad entre las muestras?
### a desviación estándar es σ = 1.22 células anormales, lo que representa aproximadamente el 58% del valor esperado (1.22 / 2.10). Esto indica una variabilidad moderada-alta: las muestras pueden alejarse bastante del promedio.

## • ¿Qué indica la función acumulada sobre la probabilidad de encontrar un número bajo o alto de células anormales?
