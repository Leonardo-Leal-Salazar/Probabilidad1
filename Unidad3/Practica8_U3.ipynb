• Una tienda registra la cantidad de unidades de cierto producto vendidas por semana.

• La variable 𝑿 representa el número de artículos vendidos, con la siguiente distribución

basada en observaciones pasadas:

Unidades vendidas (X) 0 1 2 3 4 5

Probabilidad P(X) 0.05 0.10 0.25 0.30 0.20 0.10

1. Verifique que las probabilidades sumen 1.
2. Elabore la distribución de probabilidad de 𝑿.
3. Calcule:

  • El valor esperado 𝑬 𝑿 (numero promedio de ventas).

  • La varianza y la desviación estándar.

4. Construya la función de distribución acumulada 𝑭 𝒙 y represéntela gráficamente.
5. Interprete los resultados:

• ¿Cuántas unidades se espera vender por semana?

• ¿La demanda es estable o muy variable?

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ============================================================
# PUNTO 1: Verificar que las probabilidades sumen 1
# ============================================================

x = [0, 1, 2, 3, 4, 5]           # Unidades vendidas
p = [0.05, 0.10, 0.25, 0.30, 0.20, 0.10]  # Probabilidad de cada valor

suma = sum(p)
print("Verificación de la distribución de probabilidad:")
print(f"  Suma de todas las probabilidades P(X) = {suma}")

if round(suma, 10) == 1:
    print("  ✓ La distribución es válida: las probabilidades suman 1.")
else:
    print("  ✗ Error: las probabilidades NO suman 1.")
```

```python
# ============================================================
# PUNTO 2: Distribución de probabilidad de X
# ============================================================

# Mostrar tabla de distribución
print("Distribución de Probabilidad de X:")
print(f"{'Unidades (X)':<15} {'P(X)':<10}")
print("-" * 25)
for xi, pi in zip(x, p):
    print(f"{xi:<15} {pi:<10.2f}")

# Graficar distribución
plt.stem(x, p)
plt.title("Distribución de Probabilidad de X\n(Unidades vendidas por semana)")
plt.xlabel("Número de unidades vendidas (X)")
plt.ylabel("Probabilidad P(X)")
plt.xticks(x)
plt.grid(True)
plt.show()
```

```python
# ============================================================
# PUNTO 3: Valor esperado, varianza y desviación estándar
# ============================================================

# --- Valor esperado E(X) ---
valor_esperado = 0
print("Cálculo del Valor Esperado E(X):")
print(f"{'Unidades (X)':<15} {'P(X)':<10} {'X·P(X)':<10}")
print("-" * 35)

for xi, pi in zip(x, p):
    producto = xi * pi
    valor_esperado += producto
    print(f"{xi:<15} {pi:<10.2f} {producto:<10.2f}")

print(f"\n  E(X) = {valor_esperado:.4f} unidades por semana")

# --- Varianza Var(X) ---
e_x2 = 0
print("\nCálculo de la Varianza Var(X):")
print(f"{'Unidades (X)':<15} {'P(X)':<10} {'X²·P(X)':<10}")
print("-" * 35)

for xi, pi in zip(x, p):
    producto2 = xi**2 * pi
    e_x2 += producto2
    print(f"{xi:<15} {pi:<10.2f} {producto2:<10.2f}")

varianza = e_x2 - valor_esperado**2
desviacion_std = math.sqrt(varianza)

print(f"\n  E(X²)                     = {e_x2:.4f}")
print(f"  Var(X) = E(X²) - [E(X)]² = {e_x2:.4f} - {valor_esperado:.4f}² = {varianza:.4f}")
print(f"  Desviación estándar σ = √{varianza:.4f} = {desviacion_std:.4f} unidades")
```

```python
# ============================================================
# PUNTO 4: Función de Distribución Acumulada F(x)
# ============================================================

# Calcular FDA
F = np.cumsum(p)

# Mostrar tabla
print("Función de Distribución Acumulada F(x):")
print(f"{'Unidades (X)':<15} {'P(X)':<10} {'F(X)':<10}")
print("-" * 35)
for xi, pi, fi in zip(x, p, F):
    print(f"{xi:<15} {pi:<10.2f} {fi:<10.2f}")

# Graficar
plt.step(x, F, where='post', color='steelblue', linewidth=2, label='F(x)')
plt.plot(x, F, 'o', color='steelblue', markersize=7)
plt.title("Función de Distribución Acumulada F(x)\n(Unidades vendidas por semana)")
plt.xlabel("Número de unidades vendidas (X)")
plt.ylabel("Probabilidad acumulada F(X)")
plt.xticks(x)
plt.grid(True)
plt.ylim(0, 1.05)
plt.legend()
plt.show()
```

# 5. Interprete los resultados:

## • ¿Cuántas unidades se espera vender por semana?
El valor esperado es E(X) = 2.90 unidades por semana. Esto significa que si la tienda registrara sus ventas durante muchas semanas, en promedio vendería aproximadamente 3 unidades por semana.

## • ¿La demanda es estable o muy variable?
La desviación estándar es σ = 1.30 unidades, con un coeficiente de variación del 44.8% (1.30 / 2.90 × 100). Esto indica una variabilidad moderada, lo que significa que la demanda es relativamente estable. La mayoría de las semanas las ventas se concentran entre 2 y 4 unidades, lo cual se confirma con la FDA:

El 70% de las semanas se venden 3 unidades o menos

El 90% de las semanas se venden 4 unidades o menos

Solo el 10% de las semanas se llegan a vender 5 unidades
