```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
plt.figure(figsize=(6, 4))

# Parametros de la distribucion normal
media = 70 # Media
desv_std = 4 # Desviacion estandar
total_estudiantes = 500

# VARIABLE para Calculo de La Probabilidad de que pese menos de 75 kg
peso_limite = 75

# Primero necesitamos calcular la probabilidad de que un estudiante pese menos
# de 75 kg utilizamos la funcion de distribucion acumulativa (CDF) de la detribucion normal
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)

# Ahora calculamos la probabilidad de que un estudiante pese mas de 75 kg
prob_mas_75kg = 1 - prob_menos_75kg

# Calculo del numero de estudiantes que pesan menos de 75 kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75 kg: ",prob_menos_75kg)
print("Numero de estudiantes que pesan menos de 75 kg: ", estudiantes_menos_75kg)

# Graficar la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
