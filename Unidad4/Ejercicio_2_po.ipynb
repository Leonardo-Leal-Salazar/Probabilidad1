```python
from scipy.stats import poisson

# Parametros de la distribucion de Poisson
media_muertes = 5000 * 0.001 # 5000 hombres con la probabilidad de muerte de 0.001

# Numero de reclamaciones
x = 4

# Calcular la probabilidad de que la compañia tenga que pagar 4 reclamaciones en un año determiando
probabilidad_4_reclamaciones = poisson.pmf(x, media_muertes)

print(f"La probabilidad de que la compañia tenga que pagar 4 reclamaciones es DE:",
      probabilidad_4_reclamaciones)
```
