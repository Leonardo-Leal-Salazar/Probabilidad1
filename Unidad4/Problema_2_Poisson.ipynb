El problema mismo lo indica, y cumple las condiciones: cuenta eventos (personas ingresadas) en un intervalo fijo de tiempo (un día), con una tasa media conocida (λ = 5). No hay un número fijo de ensayos.

```python
from scipy.stats import poisson

lam = 5  # media (λ)

# a) P(X = 2)
prob_a = poisson.pmf(2, lam)
print(f"a) P(X = 2) = {prob_a:.6f}  ({prob_a*100:.4f}%)")

# b) P(X > 10)
prob_b = 1 - poisson.cdf(10, lam)
print(f"\nb) P(X > 10) = {prob_b:.6f}  ({prob_b*100:.4f}%)")
print("   ¿Es probable que x exceda 10?")
if prob_b < 0.05:
    print(f"   NO es probable. Solo hay un {prob_b*100:.2f}% de chance,")
    print("   lo cual es muy bajo (menos del 5%).")
else:
    print(f"   SÍ es relativamente probable ({prob_b*100:.2f}%).")
```
