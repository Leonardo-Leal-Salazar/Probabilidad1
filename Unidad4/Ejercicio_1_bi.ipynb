```python
from scipy.stats import binom

# Parametros de la distribucion binomial
probabilidad_exito = 0.83
num_pacientes = 16
x = 16 # Numero de exitos

# Calcular la probabilidad de que ningun paciente sufra la enfermedad
probabilidad_ningun_paciente = 1 - binom.cdf(x - 1, num_pacientes, probabilidad_exito)

print("Probabilidad de que ningun paciente sufra la enfermedad:", probabilidad_ningun_paciente)
```

```python
# Parametros de la distribucion binomial
probabilidad_exito = 0.83
num_pacientes = 16
x = 0

# Calcular la probabilidad de que todos los pacientes sufran la enfermedad
probabilidad_ningun_paciente = binom.cdf(x, num_pacientes, probabilidad_exito)

print("Probabilidad de que todos los pacientes sufran la enfermedad:", probabilidad_ningun_paciente)
```
