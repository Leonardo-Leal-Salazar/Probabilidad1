 Hay un número fijo de ensayos (n = 1000 motores), cada motor tiene exactamente dos resultados posibles (defectuoso o no), y la probabilidad de éxito es constante (p = 0.001). Eso define una Binomial clásica.

```python
from scipy.stats import binom

n = 1000
p = 0.001

# P(X = 0) — ninguno defectuoso
prob_0 = binom.pmf(0, n, p)
print(f"a) P(X = 0) — ninguno defectuoso:  {prob_0:.6f}  ({prob_0*100:.4f}%)")

# P(X = 3) — tres defectuosos
prob_3 = binom.pmf(3, n, p)
print(f"b) P(X = 3) — tres defectuosos:    {prob_3:.6f}  ({prob_3*100:.4f}%)")

# P(X = 4) — cuatro defectuosos
prob_4 = binom.pmf(4, n, p)
print(f"c) P(X = 4) — cuatro defectuosos:  {prob_4:.6f}  ({prob_4*100:.4f}%)")
```
