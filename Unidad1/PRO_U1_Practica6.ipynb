```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
autos = pd.read_csv('autos2.csv')
```

```python
autos.head()
```

```python
autos.info()
```

```python
precio_min    = autos['price'].min()
precio_max    = autos['price'].max()
precio_moda   = autos['price'].mode()[0]
precio_media  = autos['price'].mean()
precio_mediana = autos['price'].median()

print(f"Valor mínimo:  ${precio_min:,.2f}")
print(f"Valor máximo:  ${precio_max:,.2f}")
print(f"Moda:          ${precio_moda:,.2f}")
print(f"Media:         ${precio_media:,.2f}")
print(f"Mediana:       ${precio_mediana:,.2f}")
```

```python
precio_std = autos['price'].std()
print(f"Desviación estándar: ${precio_std:,.2f}")

estadisticos = autos['price'].describe()
print("\nEstadísticos completos:")
print(estadisticos)
```

```python
import numpy as np

# Regla de Sturges
n = autos['price'].dropna().count()
bins_sturges = int(1 + np.log2(n))
print(f"Número de registros (n): {n}")
print(f"Bins según Regla de Sturges: {bins_sturges}")

# Histograma
plt.figure(figsize=(10, 6))
plt.hist(autos['price'].dropna(), bins=bins_sturges, color='steelblue', edgecolor='black')

plt.title('Distribución de Precios de Automóviles', fontsize=16, fontweight='bold')
plt.xlabel('Precio (USD)', fontsize=13)
plt.ylabel('Frecuencia', fontsize=13)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion
### Para la construcción del histograma se aplicó la Regla de Sturges para determinar el número óptimo de bins, utilizando la fórmula k = 1 + log₂(n), lo que garantiza una representación visual equilibrada sin exceso ni falta de intervalos. Se utilizó la función plt.hist() de matplotlib con color azul acero y bordes negros para facilitar la lectura de cada barra. Se agregaron títulos descriptivos tanto al gráfico como a los ejes, y se aplicó rotación a las etiquetas del eje X para mejorar la legibilidad. El resultado fue un histograma claro que permitió identificar la asimetría positiva en la distribución de precios

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='body-style', y='price', data=autos, palette='Set2')

plt.title('Relación entre Precio y Tipo de Vehículo', fontsize=16, fontweight='bold')
plt.xlabel('Tipo de Vehículo (Body Style)', fontsize=13)
plt.ylabel('Precio (USD)', fontsize=13)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion
### Se utilizó la función sns.boxplot() de seaborn con la paleta de colores "Set2" para diferenciar visualmente cada categoría de carrocería. Se configuraron títulos y etiquetas en ambos ejes para contextualizar correctamente la información. La rotación de las etiquetas del eje X permitió una mejor lectura de los tipos de carrocería. Este diagrama permitió comparar la dispersión y mediana de precios entre convertibles, hardtops, sedanes, hatchbacks y wagons, revelando diferencias significativas entre segmentos de mercado

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=autos, palette='Set1')

plt.title('Relación entre Precio y Tipo de Tracción', fontsize=16, fontweight='bold')
plt.xlabel('Tipo de Tracción (Drive Wheels)', fontsize=13)
plt.ylabel('Precio (USD)', fontsize=13)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion
### Se empleó la paleta "Set1" para distinguir los tres tipos de tracción (FWD, RWD y 4WD). Al igual que en el gráfico anterior, se añadieron títulos y etiquetas descriptivas, y se ajustó el tamaño de la figura para una visualización cómoda. Este diagrama evidenció que la tracción trasera (RWD) está asociada a precios más elevados, mientras que la tracción delantera (FWD) predomina en el segmento económico, confirmando que el tipo de tracción es un factor relevante en la determinación del precio.


```python
plt.figure(figsize=(9, 6))
sns.boxplot(x='engine-location', y='price', data=autos, palette='coolwarm')

plt.title('Relación entre Precio y Ubicación del Motor', fontsize=16, fontweight='bold')
plt.xlabel('Ubicación del Motor (Engine Location)', fontsize=13)
plt.ylabel('Precio (USD)', fontsize=13)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion
### Para este gráfico se utilizó la paleta "coolwarm", que por su contraste entre colores fríos y cálidos resulta ideal para comparar dos categorías opuestas como lo son el motor delantero y el motor trasero. Se mantuvieron los mismos estándares de títulos y etiquetas aplicados en los gráficos anteriores. Este fue el diagrama que mostró la diferencia de precios más marcada entre grupos, destacando que los vehículos con motor trasero pertenecen casi exclusivamente al segmento de lujo y alto rendimiento.
