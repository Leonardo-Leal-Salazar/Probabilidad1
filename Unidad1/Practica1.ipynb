```python
from statistics import *
```

```python
# Lista de calificaciones del grupo a de probabilidad y estadistica
grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
```

```python
# la funcion type nos devuelve el tipo de dato de la variable
type(grupoa)
```

```python
# Sacar la medioa de las calificaciones con una formula
promedio = sum(grupoa)/ len(grupoa)
print(promedio)
```

```python
# Utilizar la funcion mean que esta en la libreria statistics para sacar
# el promedio
promedio2 = mean(grupoa)
print(promedio2)
```

```python
# Determinar el valor de la mediana con la funcion median
mediana = median(grupoa)
print(mediana)
```

```python
# Determinar el valor de la moda con la funcion mode
moda = mode(grupoa)
print(moda)
```

```python
# En caso de que exista mas de una moda, se puede emplear la funcion
# multimode para seber las multiples modas que pueden existir
moda2 = multimode(grupoa)
print(moda2)
```

```python
# Definimos una variable llamada tiempos100mts de tipo lista, que
# contiene los tiempos que tardaron los 10 alumnos para correr 100 metros
tiempos100mts = [15.10, 17.20, 14.69, 13.27, 22.15, 18.71, 19.15, 20.65,
                 15.10, 17.20]

# Desplegamos los valores de la lista
print(tiempos100mts)
```

```python
# Sacar las medidas de tendencia
print("El tiempo promedio de los corredores es = ", mean(tiempos100mts))
print("La mediana de los tiempos de los corredores es = ", median(tiempos100mts))
print("La moda de los tiempos de los corredores es = ", mode(tiempos100mts))
```
